### Activity 9: Finding "Dream" Sentences in Les_Brown.txt

From the `Les_Brown.txt` file, read the content and identify all sentences that contain the word "dream". Write these sentences to a new file called `dream_sentences.txt`.

**Steps:**
1. Open and read the content of `Les_Brown.txt`.
2. Split the text into sentences.
3. Filter out sentences that contain the word "dream" (case-insensitive).
4. Write these sentences into the new file `dream_sentences.txt`.
<hr>

`````python
import re
#opening file
with open("Les_Brown.txt", "r") as f:
    text = f.read()
#dividing text into seperate sentences          
sentences = re.findall(r"[^.!?]+[.?!]", text)

dream = []
#loop for finding dream word in a sentence
for i in range(len(sentences)):
    if re.search("dream", sentences[i], flags=re.IGNORECASE):
        dream += sentences[i]
#Creating a new file
with open("dream_sentences.txt", "w") as f:
    f.writelines(dream)

with open("dream_sentences.txt", "r") as f:
    text = f.read()
print(text)
`````

### Activity 17: Counting Words in Les_Brown.txt

Read the content from `Les_Brown.txt`, and then count the number of occurrences of each word in the text. Write the word count results to a new file called `word_count.txt`.

**Steps:**
1. Open and read the contents of `Les_Brown.txt`.
2. Split the content into words and count the frequency of each word.
3. Write the word count in the format: `word: count` to a new file called `word_count.txt`.

<hr>

`````python
import re
#opening file
with open("Les_Brown.txt", "r") as f:
    text = f.read()
#removing everything, that is not letter
text = re.split(r"[ .,!?:\n\"]+", text)

word_count = {}
#loop for adding new words, and word count
for word in text:
    word = word.lower()
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
#writing the dictionary of words into a new file
with open("word_count.txt", "w") as f:
    for key, value in word_count.items():
        f.write(f"{key}: {value}\n")

with open("word_count.txt", "r") as f:
    text = f.read()

print(text)
`````

`````python
#Student Name: Stepan Pelc
#MISIS: M01087679

#Excellent 21

#I chose a dictionary, because all the data is very easy to see and looks clear. Also, you can add or delete data very easily.
#You can not have multiple, same key values in a dictionary, but values can be repeated, which is good for storing data like this.

dictionary = {
    "Food": [25, 40, 15],
    "Entertainment": [50, 20],
    "Travel": [300, 150]
}
#function for displaying total spendings
def total_spendings(dictionary):
    """ returns total spendings"""
    total = 0
    #loop for adding all values to the total
    for category in dictionary.values():
        for value in category:
            total += value
    return total

#loop for the user to choose what he wants to do
while True:
    action = input("Please select your desired action:\nAdd expences(Type: \"add\")\nRemove expenses(Type: \"remove\")\nStop, and see the summary(Type: \"summary\")")
    #conditions based on user input
    if action.lower().strip() == "summary":
        break
    elif action.lower().strip() == "add":
        #choosing category from dictionary
        category = input(f"Please select a category: {dictionary.keys()}, or create a new one:")
        #trying to see if the category already exists or not
        try:
            dictionary[category]
        except KeyError as ke:
            #making sure the user typed the new category name correctly
            category_new = input(f"To confirm creation of a new category: {category.capitalize()}, type (\"yes\"), to cancel type (\"no\")")
            if category_new.lower().strip() == "yes":
                value = input("Please enter your expenses:")
                #making sure input is an integer
                try:
                    value = int(value)
                except ValueError as ve:
                    print("Invalid value.")
                #adding new key and value into the dictionary
                else:
                    category = category.capitalize()
                    dictionary[category] = [value]
        else:
            value = input("Please enter your expenses:")
            #making sure input is an integer
            try:
                value = int(value)
            except ValueError as ve:
                print("Invalid value.")
            else:
                dictionary[category].append(value)
    elif action.lower().strip() == "remove":
        category = input(f"Please select a category: {dictionary.keys()}")
        if category in dictionary.keys():
            value = input("please enter the amount you want to remove:")
            try:
                value = int(value)
                dictionary[category].remove(value)
            except ValueError as ve:
                print("Value you entered is not in your account, or valid.")
        else:
            print("You did not enter a valid category name.")
    else:
        print("You did not enter a valid command.")

print(f"Your spendings:\n{dictionary}\nYour total spendings: {total_spendings(dictionary)}")
        
`````
