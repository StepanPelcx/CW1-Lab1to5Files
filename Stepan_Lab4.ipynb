````python
from string import ascii_lowercase

text = "helloworld"
key = 7

alphabet = ascii_lowercase
print(alphabet)
#loop for finding each char. in a string
for character in text:
    alphabet_char_index = alphabet.index(character)
    new_char_index = (alphabet_char_index + key) % len(alphabet) # Wrapping around 
    encoded_char = alphabet[new_char_index]
    print(f"{character} -> {alphabet_char_index} - {encoded_char}")
#loop for encoded letters
for letter in text:
    ascii_value = ord(letter)
    shift_ascii = ((ascii_value + key) % 25) + 65
    encoded_letter = chr(shift_ascii)
    print(f"{letter} -> {encoded_letter}")
````

3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
def ascii_convertor(password):
    """Returns a converted value to ASCII"""
    ascii_numbers = ""
    #converting each character in a string to ASCII numbers
    for i in range(len(password)):
        ascii_numbers += f"{ord(password[i])}, "
    return ascii_numbers

password = input("Please enter your password:")

print(ascii_convertor(password))
````

13. Secret Word Generator with Randomization: Ask the user for their first name and surname. Create a function that generates a special secret word that:
Uses the first 3 letters of the first name and last 2 letters of the surname.
Randomly appends a special character (e.g., @, #, $, etc.) at the end of the secret word.

```python
special_chars = "@#$%^&*!"
```

````python
#importing necessary module
import random

#defining a function called for the creation of the secret word
def secret_word(first_name, last_name):
    """ Returns a special word created with the user name, surname, and special characters"""
    special_chars = "@#$%^&*!"
    special_chars = random.sample(special_chars, len(special_chars))
    #creating a variable for the secret word
    word = ""
    # capitalizing the first name
    first_name = first_name.capitalize()
    #Loops for finding first 3 and last 2 characters
    #adding the first 3 characters of the first name to the word
    for n in range(3):
        word += first_name[n]
    #adding the last 2 characters of the last name to the word
    for i in range(2):
        word += last_name[i - 2]
    #adding all the special characters to the word
    for i in range(len(special_chars)):
        word += special_chars[i]
    return word

#loop to make sure user inputs only alphabetic letters
while True:
    first_name = input("Please enter your first name:").strip()
    last_name = input("Please enter your last name:").strip()
    if first_name.isalpha() and last_name.isalpha() == True:
        break
    else:
        print("Please use only alphabetic letters for the name and surname.")


print(secret_word(first_name, last_name))
````
