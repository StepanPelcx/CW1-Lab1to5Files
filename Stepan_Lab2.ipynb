4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
import math
height = 0.0254 * int(input("Please enter your height in inches: "))
weight = 0.45359237 * int(input("Please enter your weight in pounds: "))
bmi = weight / pow(height, 2)

if bmi < 18.5:
    print("Underweight")
elif bmi <= 25:
     print("Normal")
elif bmi <= 30:
     print("Overweight")
else:
    print("Obese")
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
student_marks = [30, 66, 49, 89, 90, 73, 42, 75, 56, 79, 60]
std_count = len(student_marks)
sum_marks = 0

for marks in student_marks:
    sum_marks = sum_marks + marks
average = sum_marks / std_count

lowest_mark = min(student_marks)
highest_mark = max(student_marks)

n = 1
std_passed = 0
while n <= len(student_marks):
    if student_marks[n - 1] >= 50:
        std_passed += 1
        n += 1
    else:
        n += 1

std_failed = len(student_marks) - std_passed
                 
print(f"Average mark of this exam is {average:.2f}.\nMaximum score of the exam is: {highest_mark}\nMinimum score of the exam is: {lowest_mark}\nExam passed {std_passed} students, and {std_failed} failed.")

````


13. **Vowel or Constant**

In this exercise you will create a program that read a letter of the alphabet from the user. If the user enter a, e, i, o or u then your program should display a message indicating that the entered letter is a vowel. If the user enters y then your program should display a message indicating that sometimes y is a vowel, and sometimes y is a consonants.  Otherwise your program should display a message indicating that the letter is a constant. 


````python
letter = input("Please enter one letter: ")

match letter.lower():
    case "a" | "e" | "i" | "o" | "u":
        print(f"The letter {letter} is a vowel.")
    case "y":
        print(f"The letter {letter} is sometimes vowel, and sometimes consonant.")
    case _:
        print(f"The letter {letter} is a consonant.")
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
import random
secret_number = random.randint(1,99)
number = int(input("Please enter a number from 1 to 99: "))
tries = 0

while number != secret_number:
    if number < secret_number:
        tries += 1
        number = int(input("Secret number is higher, try again: "))
    else:
        tries += 1
        number = int(input("Secret number is smaller, try again: "))
print(f"Excellent! You guessed the number!\nNumber of tries: {tries}")
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 


---

````python
numbers = []
input_numbers = input("Please enter a positive number, or stop by typing \"done\":")

while True:
    if input_numbers.lower() == "done":  # checking if input is "done"
        break
    elif input_numbers.lstrip("-").isdigit():   #checking if number isnt negative
        if int(input_numbers) > 0:
            numbers.append(int(input_numbers))
            input_numbers = input("Please enter another number, or stop by typing \"done\":")
        else:
            input_numbers = input("Please enter only positive number, or stop by typing \"done\":")
    else:
        input_numbers = input("Please enter a number, not letters, or stop by typing \"done\":")  #checking for letters in number

if len(numbers) > 0:
    print(f"All valid numbers entered: {numbers}")
    for number in numbers:
        if number / 2 == number // 2:  #finding 1st even number
            print(f"First even number is: {number}")
            break
    else:
        print("No even numbers found.")
else:
    print("No numbers were entered.")
````

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
#asking user for number of students
number_of_std = int(input("Please enter number of students: "))
std_score = []

#loop for user to input score of each student
for n in range(1,number_of_std + 1):
    score = int(input(f"Please enter score of student number {n}: "))
    #adding score to a list
    std_score.append(score)
#highest input student score
highest = max(std_score)
#sorting the list
std_score_sorted = sorted(std_score, reverse = True)

#setting variables for both highest scores.
highest = std_score_sorted[0]
second_highest = std_score_sorted[1]

#index of the highest score
highest_index = std_score.index(highest) + 1

#making sure, the second highest score has different index number, if the score is the same as the highest.
std_score[highest_index - 1] = 0
second_highest_index = std_score.index(second_highest) + 1

print(f"Student number {highest_index} got the highest score: {highest} score.\nStudent number {second_highest_index} got the second highest score: {second_highest}")
````


21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
password = "MDX123"
guess = input("Please try to guess the password:")
# making code more simple by combining lenghts of both words.
length = max(len(password), len(guess))
accuracy = 0
# checking if each character is matched
for n in range(length):
    if guess[n] == password[n]:
        accuracy += 1

if accuracy == len(password):
    print("Password cracked!")
else:
    print(f"Your accuracy was {accuracy / len(password) * 100:.0f}%.")
````
